## 1. Fórmulas de Conversión

Para entender cómo pasar de una escala a otra, utilizamos las siguientes relaciones matemáticas basadas en los puntos de congelación y ebullición del agua:

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(Nota: $\frac{9}{5}$ es igual a 1.8)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{(°F - 32) \times 5}{9}$$



---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25 °C a Fahrenheit (Temperatura ambiente)
1.  **Identificar datos:** Tenemos $°C = 25$.
2.  **Sustituir en la fórmula:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar:** $45 + 32 = 77$.
* **Resultado:** **77 °F**.

### Ejemplo B: Convertir 98.6 °F a Celsius (Temperatura corporal promedio)
1.  **Identificar datos:** Tenemos $°F = 98.6$.
2.  **Sustituir en la fórmula:** $°C = (98.6 - 32) \times \frac{5}{9}$.
3.  **Restar:** $98.6 - 32 = 66.6$.
4.  **Multiplicar por 5 y dividir entre 9:** $66.6 \times 5 = 333$; luego $333 / 9 = 37$.
* **Resultado:** **37 °C**.
